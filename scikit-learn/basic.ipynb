# An introduction to machine learning with scikit-learn

## Loading an example dataset

```python
from sklearn import datasets
iris = datasets.load_iris()
digits = datasets.load_digits()
```

```python
digits.data
```

```python
digits.target
```

```python
digits.images[0]
```

## Learning and predicting

```python
from sklearn import svm
clf = svm.SVC(gamma=0.001, C=100) # clf = classifier
```

```python
clf.fit(digits.data[:-1], digits.target[:-1])
```

```python
clf.predict(digits.data[-1:])
```

the trained classifier identifies the last image (data) as an **8**

to check this the corresponding image can be displayed like shown bellow

```python
import matplotlib.pyplot as plt
```

```python
_, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,3))
ax.set_axis_off()
ax.imshow(digits.images[-1], cmap=plt.cm.gray_r, interpolation="nearest")
```

### Type casting

```python
import numpy as np
from sklearn import kernel_approximation

rng = np.random.RandomState(0)
X = rng.rand(10, 2000)
X
```

```python
X.shape
```

```python
X.dtype
```

```python
X = np.array(X, dtype='float32')
X.dtype
```

```python
transformer = kernel_approximation.RBFSampler()
X_new = transformer.fit_transform(X)
X_new.dtype
```
