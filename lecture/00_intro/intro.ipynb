## The Power of Sampling

Auf der einen Seite haben wir den **theoretischen** Ansatz. Wir kennen die PDF der Daten und können den Erwartungswert genau berechnen.

Auf der anderen Seite haben wir den **empirischen** Ansatz. Wir ziehen viele Stichproben aus der Verteilung und berechnen den Erwartungswert anhand dieser Stichproben.

### Beispiel aus Vorlesung:

Wir haben eine Zufallsvariable $X$ mit der PDF $p(x)$. Wir wollen wissen, was passiert, wenn wir sie quadrieren. Wie sieht die PDF von $X^2$ aus?

#### Theoretischer Ansatz
Wir berechnen die PDF von $X^2$ direkt aus der PDF von $X$.
Hierfür 

#### Empirischer Ansatz
1. Ziehen viele Stichproben $x_1, x_2, \ldots, x_N$ aus der Verteilung von $X$.
2. Berechnen der Stichproben von $X^2$: $y_i = x_i^2$ für $i = 1, \ldots, N$.
3. Schätzen der PDF von $Y = X^2$ anhand der Stichproben $y_1, y_2, \ldots, y_N$.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.uniform(0,1,100000)
y = x**2

plt.hist(x,100, density=True)
plt.show()
print("Erwartungswert von X:", np.mean(x))

plt.hist(y,100, density=True)
plt.show()
print("Erwartungswert von Y:", np.mean(y))
ysum = y.sum()
E_y = ysum / len(y)
print("Erwartungswert von Y (manuell berechnet):", E_y)
```
